```python
from bs4 import BeautifulSoup
soup = BeautifulSoup()
import requests
import csv
import pandas as pd
```

```python
req = requests.get("https://www.amazon.in/s?k=redmi+note+10+pro+max&crid=34R4XIF0RQ61M&sprefix=redmi+note+10+pro%2Caps%2C280&ref=nb_sb_ss_ts-doa-p_3_17")
soup = BeautifulSoup(req.content, "html.parser")
res = soup.title
print(res.get_text())
```

```python
product = []
prices = []
ratings = []
product = ('a-size-medium a-color-base a-text-normal')
print(product)
```

```python
for a in soup.findAll('a', href = True, attrs ={'class':'div.sg-col-inner'}):
    name=a.find('div', attrs ={'class':'a-size-medium a-color-base a-text-normal'})
    prices=a.find('div', attrs ={'class':'a-price-whole'})
    ratings=a.find('div', attrs ={'class':'a-icon a-icon-star-small a-star-small-4 aok-align-bottom'})
    product.append(name.text)
    prices.append(prices.text)
    ratings.append(ratings.text)
```

```python
df = pd.DataFrame({"Product name": products, 'Prices':prices, "Ratings":ratings})
df.head()
```

```python
df.to_csv("Products.csv")
```

```python
# Import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Copy the url of webpage
url ="https://www.flipkart.com/search?q=redmi%20note%2010%20pro%20max&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
# Parse the content into HTML
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content, 'html.parser')
```

```python
print(content)
```

```python
# Rest Code
data = content.find_all('div', {'class': '_2kHMtA'})
print(data)
```

```python
links = []
phone_names = []
start_link = "https://www.flipkart.com/"
```

```python
for items in data:
    rest_link = items.find('a')['href']
    names = items.find('div', attrs={'class':'_4rR01T'})
    phone_names.append(names.text)
    links.append(start_link+rest_link)
```

```python
print(phone_names[0])
print(links[0])
```

```python
dataframe = {'Phone_names':phone_names, 'Links':links}
Final_dataframe = pd.DataFrame(dataframe)
print(Final_dataframe)
```

```python
Final_dataframe.to_csv('FK_data_url.csv')
```

```python
import scrapy

class AmazonReviewsSpider(scrapy.Spider):

    # Spider name
    name = 'amazon_reviews'

    # Domain names to scrape
    allowed_domains = ['amazon.in']

    # Base URL for the product reviews
    myBaseUrl = "https://www.amazon.in/product-reviews/9387779262/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&pageNumber="
    start_urls=[]

    # Creating list of urls to be scraped by appending page number a the end of base url
    for i in range(1,121):
        start_urls.append(myBaseUrl+str(i))

    # Defining a Scrapy parser
    def parse(self, response):
            data = response.css('#cm_cr-review_list')
            
            # Collecting product star ratings
            star_rating = data.css('.review-rating')

            # Collecting user reviews
            comments = data.css('.review-text')
            count = 0

            # Combining the results
            for review in star_rating:
                yield{'stars': 
                      ''.join(review.xpath('.//text()').extract()),
                      'comment': 
                          ''.join(comments[count].xpath(".//text()").extract())
                     }
                count=count+1
```
